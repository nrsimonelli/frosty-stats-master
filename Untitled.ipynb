```python
def stack_per_player_data(df):

    total_wins = pd.concat([df.groupby('t1_player1')['t1_win'].sum(),
                            df.groupby('t1_player2')['t1_win'].sum(),
                            df.groupby('t1_player3')['t1_win'].sum(),
                            df.groupby('t1_player4')['t1_win'].sum(),
                            df.groupby('t2_player1')['t2_win'].sum(),
                            df.groupby('t2_player2')['t2_win'].sum(),
                            df.groupby('t2_player3')['t2_win'].sum(),
                            df.groupby('t2_player4')['t2_win'].sum()], axis=1).sum(axis=1)

    total_points = pd.concat([df.groupby('t1_player1')['t1_points'].sum(),
                            df.groupby('t1_player2')['t1_points'].sum(),
                            df.groupby('t1_player3')['t1_points'].sum(),
                            df.groupby('t1_player4')['t1_points'].sum(),
                            df.groupby('t2_player1')['t2_points'].sum(),
                            df.groupby('t2_player2')['t2_points'].sum(),
                            df.groupby('t2_player3')['t2_points'].sum(),
                            df.groupby('t2_player4')['t2_points'].sum()], axis=1).sum(axis=1)

    total_points_played = pd.concat([df.groupby('t1_player1')['total_points'].sum(),
                                     df.groupby('t1_player2')['total_points'].sum(),
                                     df.groupby('t1_player3')['total_points'].sum(),
                                     df.groupby('t1_player4')['total_points'].sum(),
                                     df.groupby('t2_player1')['total_points'].sum(),
                                     df.groupby('t2_player2')['total_points'].sum(),
                                     df.groupby('t2_player3')['total_points'].sum(),
                                     df.groupby('t2_player4')['total_points'].sum()], axis=1).sum(axis=1)

    total_games = pd.concat([df.groupby('t1_player1').size(),
                             df.groupby('t1_player2').size(),
                             df.groupby('t1_player3').size(),
                             df.groupby('t1_player4').size(),
                             df.groupby('t2_player1').size(),
                             df.groupby('t2_player2').size(),
                             df.groupby('t2_player3').size(),
                             df.groupby('t2_player4').size(),], axis=1).sum(axis=1)


    tbl = pd.concat([total_wins,total_points,total_points_played,total_games], axis=1)
    tbl.columns = ['tot_wins', 'total_points_scored', 'total_points_played','tot_games']
    tbl['win_perc'] = tbl['tot_wins'] / tbl['tot_games']
    tbl['ppg'] = tbl['total_points_scored'] / tbl['tot_games']
    tbl['point_score_perc'] = tbl['total_points_scored'] / tbl['total_points_played']
    return tbl
```

```python
tbl = stack_per_player_data(df)
tbl.sort_values('win_perc', ascending=False)
```
